```python
import os.path as op
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pipeline='NEARICA_behav'
```

```python
kinematics_9m=pd.read_csv(op.join('/home/bonaiuto/dev_beta_umd/data/9m/derivatives',pipeline,'processed_kinematics.csv'))
kinematics_12m=pd.read_csv(op.join('/home/bonaiuto/dev_beta_umd/data/12m/derivatives',pipeline,'processed_kinematics.csv'))
kinematics_adult=pd.read_csv(op.join('/home/bonaiuto/dev_beta_umd/data/adult/derivatives',pipeline,'processed_kinematics.csv'))
```

```python
nine_color=np.array([228,26,28])/255.0
twelve_color=np.array([55,126,184])/255.0
adult_color=np.array([77,175,74])/255.0

plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
_=plt.hist(kinematics_9m['ReachDur'][kinematics_9m['Condition']=='exe'],300,alpha=.5,density=True,color=nine_color,label='9m')
_=plt.hist(kinematics_12m['ReachDur'][kinematics_12m['Condition']=='exe'],300,alpha=.5,density=True,color=twelve_color,label='12m')
_=plt.hist(kinematics_adult['ReachDur'][kinematics_adult['Condition']=='exe'],100,alpha=.5,density=True,color=adult_color,label='adult')
plt.xlim([0,25])
plt.xlabel('Time to reach (s)')
plt.legend()

plt.subplot(2,3,2)
_=plt.hist(kinematics_9m['GraspDur'][kinematics_9m['Condition']=='exe'],40,alpha=.5,density=True,color=nine_color)
_=plt.hist(kinematics_12m['GraspDur'][kinematics_12m['Condition']=='exe'],40,alpha=.5,density=True,color=twelve_color)
_=plt.hist(kinematics_adult['GraspDur'][kinematics_adult['Condition']=='exe'],40,alpha=.5,density=True,color=adult_color)
plt.xlim([0,1.6])
plt.xlabel('Grasp duration (s)')

plt.subplot(2,3,3)
_=plt.hist(kinematics_9m['EndDur'][kinematics_9m['Condition']=='exe'],200,alpha=.5,density=True,color=nine_color)
_=plt.hist(kinematics_12m['EndDur'][kinematics_12m['Condition']=='exe'],200,alpha=.5,density=True,color=twelve_color)
_=plt.hist(kinematics_adult['EndDur'][kinematics_adult['Condition']=='exe'],200,alpha=.5,density=True,color=adult_color)
plt.xlim([0,5])
plt.xlabel('Manipulation duration (s)')

plt.subplot(2,3,4)
_=plt.hist(kinematics_9m['ReachDur'][kinematics_9m['Condition']=='obs'],300,alpha=.5,density=True,color=nine_color)
_=plt.hist(kinematics_12m['ReachDur'][kinematics_12m['Condition']=='obs'],200,alpha=.5,density=True,color=twelve_color)
_=plt.hist(kinematics_adult['ReachDur'][kinematics_adult['Condition']=='obs'],200,alpha=.5,density=True,color=adult_color)
plt.xlim([0,25])
plt.xlabel('Time to reach (s)')

plt.subplot(2,3,5)
_=plt.hist(kinematics_9m['GraspDur'][kinematics_9m['Condition']=='obs'],40,alpha=.5,density=True,color=nine_color)
_=plt.hist(kinematics_12m['GraspDur'][kinematics_12m['Condition']=='obs'],40,alpha=.5,density=True,color=twelve_color)
_=plt.hist(kinematics_adult['GraspDur'][kinematics_adult['Condition']=='obs'],40,alpha=.5,density=True,color=adult_color)
plt.xlim([0,1.6])
plt.xlabel('Grasp duration (s)')

plt.subplot(2,3,6)
_=plt.hist(kinematics_9m['EndDur'][kinematics_9m['Condition']=='obs'],20,alpha=.5,density=True,color=nine_color)
_=plt.hist(kinematics_12m['EndDur'][kinematics_12m['Condition']=='obs'],100,alpha=.5,density=True,color=twelve_color)
_=plt.hist(kinematics_adult['EndDur'][kinematics_adult['Condition']=='obs'],150,alpha=.5,density=True,color=adult_color)
plt.xlim([0,5])
plt.xlabel('Manipulation duration (s)')
```
